```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# --------------------------
# QUESTIONS (page 3)
# --------------------------
questions = [
    ("1. Šis asmuo kalba…", ["tyliai", "garsiai", "užtikrintai", "nerodydamas emocijų"]),
    ("2. Kalbėdamas šis asmuo…", ["ramiai", "labai greitai", "pertraukia", "kontroliuoja pokalbį"]),
    ("3. Klausydamas šis asmuo…", ["atsako greitai", "draugiškas", "energingas", "atidžiai klauso"]),
    ("4. Užduodamas klausimus…", ["apie jausmus", "esminiai faktai", "daug info", "kaip klausimai"]),
    ("5. Elgsena…", ["daug juda", "gestikuliuoja", "santūrus", "akių kontaktas"]),
    ("6. Reakcija…", ["entuziastingas", "atsipalaidavęs", "nerimastingas", "išlaiko atstumą"]),
    ("7. Bendravimas…", ["draugiškas", "stebi", "vadovauja", "reiškia jausmus"]),
    ("8. Į užduotis…", ["spontaniškas", "greitas sprendimas", "atsargus", "sistemingas"])
]

# --------------------------
# KEY (page 8)
# order: [option1, option2, option3, option4]
# --------------------------
key = [
    ["S","i","D","C"],
    ["S","i","C","D"],
    ["D","S","i","C"],
    ["i","D","C","S"],
    ["D","i","C","S"],
    ["i","S","D","C"],
    ["S","C","D","i"],
    ["i","D","C","S"]
]

# --------------------------
# UI STORAGE
# --------------------------
labiausiai = []
maziausiai = []

# --------------------------
# BUILD UI
# --------------------------
ui = []

for i, (q, opts) in enumerate(questions):
    lab = widgets.RadioButtons(options=opts, description="Labiausiai:")
    maz = widgets.RadioButtons(options=opts, description="Mažiausiai:")

    def make_observer(lab, maz):
        def handler(change):
            if lab.value == maz.value and lab.value is not None:
                maz.value = None  # prevent same selection
        return handler

    lab.observe(make_observer(lab, maz), names='value')
    maz.observe(make_observer(lab, maz), names='value')

    labiausiai.append(lab)
    maziausiai.append(maz)

    ui.append(widgets.HTML(f"<b>{q}</b>"))
    ui.append(widgets.HBox([lab, maz]))

submit_btn = widgets.Button(description="Skaičiuoti rezultatą", button_style='success')
output = widgets.Output()

# --------------------------
# CALCULATION
# --------------------------
def calculate(b):
    with output:
        clear_output()

        counts = {"D":0, "i":0, "S":0, "C":0}
        total = 0

        for i in range(len(questions)):
            lab = labiausiai[i].value
            maz = maziausiai[i].value

            if lab is None or maz is None:
                print(f"❌ Neatsakyta į klausimą {i+1}")
                return

            opts = questions[i][1]

            lab_index = opts.index(lab)
            maz_index = opts.index(maz)

            counts[key[i][lab_index]] += 1
            counts[key[i][maz_index]] += 1

            total += 2

        if total != 16:
            print("❌ Klaida: suma turi būti 16")
            return

        print("Rezultatas:")
        print(f"D-{counts['D']}, i-{counts['i']}, S-{counts['S']}, C-{counts['C']}")


submit_btn.on_click(calculate)

display(*ui, submit_btn, output)
```
